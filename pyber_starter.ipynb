```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data
city_df = pd.read_csv(city_data_to_load)
ride_df = pd.read_csv(ride_data_to_load )

# Combine the data into a single dataset
combined_df = pd.merge(ride_df, city_df, 
                                 how='outer',on='city')

# Display the data table for preview
combined_df.head()
```

## Bubble Plot of Ride Sharing Data

```python
combined_df["color"] =  "coral"

combined_df.loc[combined_df['type']  == 'Suburban',  'color'] = 'lightskyblue'
combined_df.loc[combined_df['type']  == 'Rural',  'color'] = 'gold'
```

```python
# Obtain the x and y coordinates for each of the three city types
#Number_of_rides = combined_df.groupby("ride_id").count()
#Average_fare = combined_df.groupby("fare")

#types = combined_df['type'].unique



group_by_city = combined_df.groupby("city")
Number_of_rides = group_by_city["ride_id"].count()
Number_of_rides

Average_fare = group_by_city["fare"].mean()
Average_fare

driver_per_city = group_by_city["driver_count"].sum()
driver_per_city

colors_list = group_by_city["color"].unique()
colors = [color[0] for color in colors_list]

#lightskyblue, coral, yellow
#colors = group_by_city['type' == "Urban", color = 'blue']
#colors2 = group_by_city('type' == "Suburban", color= 'red')
#colors3 = group_by_city('type' == "Rural", color= 'yellow')

#color = ['red'== Number_of_type("Urban"), 'blue'== Number_of_type("Suburban"), 'yellow'==Number_of_type("Rural")]


#colors = group_by_city["type"].count


# Build the scatter plots for each city types
plt.scatter(Number_of_rides, Average_fare, marker="o", facecolors=colors, edgecolors="black", 
            s = driver_per_city, alpha =0.75)
plt.grid()
# Incorporate the other graph properties

# Create a legend
plt.title("Pyber Ride Sharing Data (2016)")
plt.xlabel("Total Number of Rides (Per City)")
plt.ylabel("Average Fare ($)")


# Incorporate a text label regarding circle size

# Save Figure
plt.savefig("Pyscatter.png")
plt.show()
```

## Total Fares by City Type

```python
# Calculate Type Percents
total_fares = combined_df["fare"].sum() 

group_by_type = combined_df.groupby("type")
fare_per_type = (group_by_type["fare"].sum()/total_fares)*100
fare_per_type

colors = ["gold","lightskyblue", "lightcoral"]

explode = (0,0,0.1)

# Build Pie Chart
plt.pie(fare_per_type, labels= ["Rural", "Suburban", "Urban"], explode=explode, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=140)

plt.title("% of Total Fares by City Type")

# Save Figure
plt.savefig("PyPie1.png")
plt.show()
```

## Total Rides by City Type

```python
# Calculate Ride Percents

total_count = combined_df["ride_id"].count() 

group_by_type = combined_df.groupby("type")
ride_per_type = (group_by_type["ride_id"].count()/total_count)*100
ride_per_type

colors = ["gold","lightskyblue", "lightcoral"]

explode = (0,0,0.1)

# Build Pie Chart
plt.pie(ride_per_type, labels= ["Rural", "Suburban", "Urban"], explode=explode, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=140)

plt.title("% of Total Rides by City Type")

# Save Figure
plt.savefig("PyPie2.png")
plt.show()

# Build Pie Chart

# Save Figure
```
